```python
def shell_sort(arr):
    
    #n/2,n/4,n/8.....
    n = len(arr)
    interval = n//2
    
    ##decrease till gap == 1, once gap == 1, its just a normal insertion sort
    while interval > 0:
        
        for i in range(interval,n):
            #perform insertion sort
            currentval = arr[i]
            position = i
            
            while position >= interval and arr[position-interval] > currentval:
                arr[position] = arr[position-interval]
                position -= interval
                
            arr[position] = currentval
        
        #decrease the gap
        interval = interval//2
        
    return arr
```

```python
shell_sort([90,1,3,2,100,2029,100,90,1,3000])
```

```python
import random
arr = []
for a in range(1000):
    arr.append(random.randint(0,1000))
```

```python
%%time

shell_sort(arr)
```
